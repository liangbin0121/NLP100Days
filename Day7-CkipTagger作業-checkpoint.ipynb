### 進行斷詞, 詞性標注, 與命名實體識別

```python
#創建實例
###<your code>###

#斷詞
###<your code>###
# 使用Ckiptagger進行斷詞
ws = WS("../ckip_data") #建構斷詞
word_s = ws(
    sentence_list,
    sentence_segmentation = True, # To consider delimiters
    segment_delimiter_set = {",", "。", ":", "?", "!", ";"}) # This is the defualt set of delimiters

print(f'斷詞輸出: {word_s}')
print('\n')

#詞性標注
###<your code>###
# 使用Ckiptagger進行詞性標註
pos = POS("../ckip_data")

word_p = pos(sentence_list)
print(f'詞性標注輸出: {word_p}')
print('\n')
```

```python

#命名實體識別
###<your code>###
# 使用Ckiptagger進行命名實體辨識
ner = NER("../ckip_data")

word_n = ner(word_s, word_p)
print(f'命名實體識別輸出: {word_n}')
```
