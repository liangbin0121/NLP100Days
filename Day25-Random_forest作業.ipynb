### 作業目的: 了解何謂Bias-Variance Tradeoff

### Q1: 一般模型誤差可拆為三部分，這三部分的誤差分別是？

Answer: trainning error/validation error/test error

### Q2: 請解釋何謂Bias-Variance Tradeoff

Answer: 模式的整體誤差(Total error)=偏差(Bias)+變異(variance)+隨機誤差(random error)

有兩種問題:
Overfitting (模式複雜度過高): Bias降低，但Variance增加

Underfitting (模式複雜度過低): Variance降低，但Bias增加

我們不希望模型有Overfitting和Underfitting，無法同時使bias和variance降低，只能採折衷方法
