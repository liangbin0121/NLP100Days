### 取出訓練特徵與標註

```python
X = dataset[['User ID', 'Gender', 'Age', 'EstimatedSalary']].values
Y = dataset['Purchased'].values
```

---

```python
import numpy as np
from sklearn.model_selection import KFold
```

### 將訓練資料按照順序切割成10等分

```python
kf = KFold(n_splits=10, shuffle=False, random_state=None)
#kf = KFold("自行填入")

kf.get_n_splits(X)

print(kf)
```

### 將訓練資料隨機切割成10等分

```python
kf = KFold(n_splits=10, shuffle=True, random_state=None)
#kf = KFold("自行填入")
kf.get_n_splits(X)

print(kf)
```

---

### 取出 切割資料對應位置

```python
train_split = kf.split(X)
next(train_split)
```

### Or

```python
for train_index, test_index in kf.split(X):
    print("TRAIN:", train_index, "TEST:", test_index)
```

### 取出切割資料：trainset / testset 特徵(x_train/x_test)/標註(y_train/y_test)

```python
kf_dataset=[]
X_train_dataset=[]
Y_train_dataset=[]
X_test_dataset=[]
Y_test_dataset=[]
for train_index, test_index in kf.split(X):
  kf_dataset.append((train_index, test_index))
  X_train_dataset.append(X[train_index])
  X_test_dataset.append(X[test_index])
  Y_train_dataset.append(Y[train_index])
  Y_test_dataset.append(Y[test_index])
  #X_train, X_test = X[train_index], X[test_index]
  #y_train, y_test = y[train_index], y[test_index]
```
